```python
column_counts = df.shape[1]  
data_points = df.shape[0]    
null_columns = df.columns[df.isnull().any()]  
dtype_counts = df.dtypes.value_counts() 

print("Number of columns:", column_counts)
print("Number of data points:", data_points)
print("Columns with null values:", null_columns.tolist())
print("Data type counts:")
print(dtype_counts)
```

```python
df['Likes'] = df['Likes'].fillna(0)
df['Comments'] = df['Comments'].fillna(0)

df.dropna(inplace=True)
```

```python
album_type_counts = df['Album_type'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(album_type_counts, labels=album_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Album Types')
plt.axis('equal')  
plt.show()

print("Remarks:")
print("Albums make the majority, followed by singles and then compilations.")
```

```python
artist_youtube_views = df.groupby('Artist')['Views'].sum().nlargest(10)
artist_spotify_streams = df.groupby('Artist')['Stream'].sum().nlargest(10)

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
artist_youtube_views.plot(kind='bar', color='red')
plt.title('Top 10 Artists on YouTube')
plt.xlabel('Artist')
plt.ylabel('Total Views')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
artist_spotify_streams.plot(kind='bar', color='green')
plt.title('Top 10 Artists on Spotify')
plt.xlabel('Artist')
plt.ylabel('Total Streams')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

print("Remarks:")
print("In Youtube Ed Sheeran is the Top 1 Artist, in comparison to Spotify where he is only 2nd, with Post Malone Being Top 1 in Spotify.")
print("Despite the fact that Post Malone doesn't reach Top 10 on Youtube views.")
```

```python
song_views = df.groupby('Track')['Views'].sum().nlargest(10)
song_comments = df.groupby('Track')['Comments'].sum().nlargest(10)
song_likes = df.groupby('Track')['Likes'].sum().nlargest(10)

print("Top 10 Songs with Most Views:")
print(song_views)

print("\nTop 10 Songs with Most Comments:")
print(song_comments)

print("\nTop 10 Songs with Most Likes:")
print(song_likes)

print("\nRemarks:")
print("Despacito is the Top 1 song on most views and most likes but only makes Top 4 in most comments.")
print("Dynamite is instead the Top 1 song on most comments despite not making Top 10 on most views and only Top 8 on Most Likes")
```

```python
top_songs = df.groupby('Track').agg({'Stream': 'sum', 'Energy': 'mean', 'Danceability': 'mean', 'Acousticness': 'mean'})

top_songs = top_songs.nlargest(10, 'Stream')

plt.figure(figsize=(12, 12))

plt.subplot(3, 1, 1)
plt.bar(top_songs.index, top_songs['Energy'], color='blue')
plt.title('Energy of Top 10 Songs by Stream')
plt.ylabel('Energy')
plt.xticks(rotation=45, ha='right') 

plt.subplot(3, 1, 2)
plt.bar(top_songs.index, top_songs['Danceability'], color='purple')
plt.title('Danceability of Top 10 Songs by Stream')
plt.ylabel('Danceability')
plt.xticks(rotation=45, ha='right') 

plt.subplot(3, 1, 3)
plt.bar(top_songs.index, top_songs['Acousticness'], color='orange')
plt.title('Acousticness of Top 10 Songs by Stream')
plt.ylabel('Acousticness')
plt.xticks(rotation=45, ha='right') 

plt.subplots_adjust(hspace=1.5)
plt.show()

print("\nRemarks:")
print("Can't Hold us is the most Energetic Song, Sunflower and Se침orita are about equal in Danceability")
print("And lovely is the most Acoustic")
```
